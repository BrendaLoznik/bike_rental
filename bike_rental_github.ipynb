I have a feeling that the Zero windspeed days are days in which the windspeed was not recorded properly. Let's see what MICE imputer can do for us here.

MICE WINDSPEED

```
#MICE imputing

#create a copy of the dataframe
mice_imputed = no_outliers.copy()

#change the data of zero windspeed days to NaN
mask = (mice_imputed ["windspeed"] == 0)
mice_imputed .loc[mask, "windspeed"] = np.nan

#select the categorical features for one hot encoding
categorical_cols = [cname for cname in mice_imputed.columns if
                    mice_imputed[cname].dtype == "object"]

# Create a one-hot encoded set 
ohe = pd.get_dummies(mice_imputed[categorical_cols], prefix="OHE")
 
# Concatenate this set back to the DataFrame
mice_imputed_ohe = pd.concat([mice_imputed, ohe], axis=1)

#drop the un-encoded features
to_drop = categorical_cols
mice_imputed_ohe= mice_imputed_ohe.drop(to_drop, axis=1)

#split dataframe in train and test and drop the datetime feature (cannot be used for modelling)
mice_train = mice_imputed_ohe[mice_imputed_ohe["datetime"].isin(train["datetime"])]
mice_train = mice_train.drop("datetime", axis=1)
mice_test =  mice_imputed_ohe[mice_imputed_ohe["datetime"].isin(test["datetime"])]
mice_test = mice_test.drop("datetime", axis=1)

#apply mice imputer to missing values (fit on training data)
mice_imputer = IterativeImputer()
mice_train.iloc[:, :] = np.round(mice_imputer.fit_transform(mice_train))
mice_test.iloc[:, :] = np.round(mice_imputer.transform(mice_test))

#merge mice_train and mice_test
mice_imputed = pd.concat([mice_train, mice_test])
```

```
#Create Density plots: the peak at 0 windspeed is gone, we find more data in the 0 - 20 range
raw['windspeed'].plot(kind='kde', c='pink', linewidth=1)
mice_imputed['windspeed'].plot(kind='kde', c = "green", linewidth = 1)
labels = ['Baseline',  'MICE']
plt.legend(labels)
plt.xlabel('windspeed')
plt.show()
```

```
df.head()
```

```
df_train = df[df["datetime"].isin(train["datetime"])]
df_test =  df[df["datetime"].isin(test["datetime"])]
```

```
#define X and y
X = df_train.drop(["datetime","count", "countlog", "windspeed_mice", "registered", "casual" ], axis=1)
y = np.log1p(df_train["count"])

# Create training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=42)
```

## **4.1 TESTING DIFFERENT MODELS**

```
models=[LinearRegression(), Ridge(), RandomForestRegressor(),AdaBoostRegressor(),BaggingRegressor(),SVR(),KNeighborsRegressor(), 
        XGBRegressor(), GradientBoostingRegressor()]
names =['Linear Regression', 'Ridge Regresion', 'Random Forest Regressor','AdaBoost Regressor','Bagging Regressor','SVR','KNeighbors Regressor', 'XGBoost Regressor', 'Gradient Boosting Regressor']
rmsle=[]
d={}
for model in range (len(models)):
    clf=models[model]
    clf.fit(X_train,y_train)
    test_pred=clf.predict(X_test)
    test_pred_exp = np.expm1(test_pred)
    y_test_exp = np.expm1(y_test)
    rmsle.append(np.sqrt(mean_squared_log_error(y_test_exp , test_pred_exp)))
d={'Model':names,'RMSLE':rmsle}   
rmsle_df = pd.DataFrame(d)
rmsle_df
```

The Random Forest Regressor, Bagging Regressor, XGB Regressor and Gradient Boosting Regressor have the lowest scores, so we will tune these models a bit further. Note: I will keep use the GridSearch function after finding the optimal parameters. I could have used the best_estimator_ of the GridSearch object, but I like seeing what the model looks like....

## **4.2 RANDOM FOREST**

TUNING

```
#Let's find optimum parameters for each model
model_rf = RandomForestRegressor(n_estimators=500, random_state=42)

params ={'max_depth':[4, 5, 6],
               'max_features':['auto','sqrt', 'log2']}


#implement GridSearch
cv_rf_model = GridSearchCV(param_grid = params, estimator=model_rf, cv=5, verbose=1 )
cv_rf_model.fit(X_train, y_train)
print('Best Score:', cv_rf_model.best_score_)
print('Parameters:', cv_rf_model.best_params_)
```

TUNED

```
model_rf = RandomForestRegressor(n_estimators=500, random_state=42)
params ={'max_depth':[ 6],
         'max_features':['auto']}

#implement Grid Search
cv_rf_model = GridSearchCV(model_rf, param_grid = params , cv=5, verbose = 1, scoring = "neg_mean_squared_log_error")

# Fit cv to the training set: 
cv_rf_model.fit(X_train, y_train)
y_pred = cv_rf_model.predict(X_test)
y_pred_log = np.expm1(y_pred)
y_test_log = np.expm1(y_test)
score = np.sqrt(mean_squared_log_error(y_test_log,y_pred_log))
print("Negative mean squared log error: {}".format(score))
```

**SCORING**

Initially I got a lot of Value Errors indicating that the Mean Squared Logarithmic Error cannot be used when targets contain negative values. I found that it also struggled with zero's.  This is exactly why I couldn't use Count (as it contained zero's) and Countlog (as it contained negative values). I finally figured out that I had to use np.log1p, which pretty much takes the log of x+1. This avoids negative values, and a lot of errors.
